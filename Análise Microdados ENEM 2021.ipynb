## Indicadores Socioeconomicos

Q001 - Até que série seu pai, ou o homem responsável por você, estudou?


Q002 - Até que série sua mãe, ou a mulher responsável por você, estudou?

E agora? Vamos definir dicionários

extração de informações

```python
q001e002Dicionario = {'A': 'Nunca estudou', 'B': 'Não completou a 4ª série/5º ano do Ensino Fundamental.', 'C' : 'Completou a 4ª série/5º ano, mas não completou a 8ª série/9º ano do Ensino Fundamental.', 'D': 'Completou a 8ª série/9º ano do Ensino Fundamental, mas não completou o Ensino Médio.', 'E': 'Completou o Ensino Médio, mas não completou a Faculdade.', 'F': 'Completou a Faculdade, mas não completou a Pós-graduação.', 'G': 'Completou a Pós-graduação.', 'H': 'Não sei.'}
```

```python
q001e002Dicionario
```

```python
microdadosEnemSelecionado['NO_Q001'] = [q001e002Dicionario[resp] for resp in microdadosEnemSelecionado.Q001]
```

```python
microdadosEnemSelecionado.head()
```

```python
microdadosEnemSelecionado['NO_Q002'] = [q001e002Dicionario[resp] for resp in microdadosEnemSelecionado.Q002]
```

```python
microdadosEnemSelecionado.head()
```

```python
microdadosEnemSelecionado.head()
```

## Comparando a distribuição do nível de escolaridade Pai x Mãe no conjunto de dados

```python
microdadosEnemSelecionado.filter(items=['NU_INSCRICAO', 'NO_Q001']).groupby('NO_Q001').count().sort_values(by = 'NU_INSCRICAO', ascending = False)
```

Distribuição dos candidatos do ENEM 2021 por escolaridade da Mãe

```python
microdadosEnemSelecionado.filter(items=['NU_INSCRICAO', 'NO_Q002']).groupby('NO_Q002').count().sort_values(by = 'NU_INSCRICAO', ascending = False)
```

## Olhando o desempenho em Matemática segundo a escolaridade do Pai

```python
microdadosEnemSelecionado.filter(items=['NU_NOTA_MT', 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by = 'NU_NOTA_MT', ascending = False)
```

## Olhando o desempenho em Matemática segundo a escolaridade da Mãe

```python
microdadosEnemSelecionado.filter(items=['NU_NOTA_MT', 'NO_Q002']).groupby('NO_Q002').mean().sort_values(by = 'NU_NOTA_MT', ascending = False)
```

O que podemos entender com esses dados? Que a média da nota de matemática aumenta conforme o grau de escolaridade tanto do pai, quanto da mãe, aumenta.

Utilização do  WHERE


```python
fig, ax = plt.subplots(figsize=(20, 10))
plt.suptitle('Nota de Redação x Escolaridade Mãe x Estado')

microdadosEnemSelecionado.filter(items=['SG_UF_PROVA', 'Q002', 'NU_NOTA_REDACAO'])\
                                .where(microdadosEnemSelecionado.Q002 != 'H')\
                                .groupby(['Q002', 'SG_UF_PROVA'])\
                                .mean().sort_values(by='NU_NOTA_REDACAO', ascending = False)\
                                .unstack().plot(ax=ax, )
```

Ou seja, se confirma a hipótese de que, mesmo por região. A medida que o grau de escolaridade dos responsáveis aumenta, a nota do candidato tende a aumentar.
